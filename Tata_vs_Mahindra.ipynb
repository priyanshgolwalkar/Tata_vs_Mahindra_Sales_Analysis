```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"D:\IdeaSpark\Tata_Mahindra_sales.xlsx"

df = pd.read_excel(file_path)

df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
monthly_sales = df.groupby(['Date', 'Brand'])['Total Sales'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Date', y='Total Sales', hue='Brand', marker='o')
plt.title('Monthly Sales Trend: Tata vs Mahindra')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_market_share = df.groupby('Brand')['Market Share (%)'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_market_share, x='Brand', y='Market Share (%)', palette='Set2')
plt.title('Average Market Share (2021-2024)')
plt.ylabel('Market Share (%)')
plt.show()
```

```python
segment_sales = df.groupby(['Segment', 'Brand'])['Total Sales'].sum().unstack().fillna(0)

segment_sales.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Paired')
plt.title('Segment-Wise Sales Comparison')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Brand')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Engine Type', hue='Brand', palette='pastel')
plt.title('Fuel Type Popularity')
plt.xlabel('Engine Type')
plt.ylabel('Number of Models')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Brand', y='Price (₹)', palette='muted')
plt.title('Price Distribution: Tata vs Mahindra')
plt.ylabel('Price (₹)')
plt.show()
```

```python
yearly_sales = df.groupby(['Year', 'Brand'])['Total Sales'].sum().reset_index()
yearly_sales['Growth Rate'] = yearly_sales.groupby('Brand')['Total Sales'].pct_change().fillna(0) * 100
yearly_sales['Year'] = yearly_sales['Year'].astype(str)

plt.figure(figsize=(10, 6))
sns.barplot(data=yearly_sales, x='Year', y='Growth Rate', hue='Brand', palette='coolwarm')
plt.title('Year-over-Year Growth Rate')
plt.ylabel('Growth Rate (%)')
plt.show()
```

```python
top_models = df.groupby("Model")["Sales Volume"].sum().nlargest(5)
top_models
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=top_models.values, y=top_models.index, palette="viridis")
plt.xlabel("Total Sales")
plt.ylabel("Car Model")
plt.title("Top 5 Best-Selling Models")
plt.show()
```

```python
# Group data by Segment and Brand, summing total sales
segment_brand_sales = df.groupby(['Segment', 'Brand'])['Total Sales'].sum().reset_index()

# Get dominant brand per segment
dominant_brand = segment_brand_sales.sort_values('Total Sales', ascending=False).groupby('Segment').first().reset_index()

# Plot the dominant brand for each segment
plt.figure(figsize=(10, 6))
sns.barplot(data=dominant_brand, x='Total Sales', y='Segment', hue='Brand', palette='Set2')

plt.title('Which Brand Dominates Each Segment')
plt.xlabel('Total Sales')
plt.ylabel('Vehicle Segment')
plt.legend(title='Brand')
plt.tight_layout()
plt.show()
```

```python
heatmap_df = df.groupby(['Year', 'Month'])['Total Sales'].sum().unstack()

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_df, annot=False, cmap='YlGnBu')  # Changed annot to False
plt.title('Heatmap of Total Sales by Year and Month')
plt.ylabel('Year')
plt.xlabel('Month')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Price (₹)', y='Market Share (%)', hue='Brand')
plt.title('Price vs Market Share')
plt.show()
```

```python
df_sorted = df.sort_values(by='Date')
df_sorted['Cumulative Sales'] = df_sorted.groupby('Brand')['Total Sales'].cumsum()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_sorted, x='Date', y='Cumulative Sales', hue='Brand')
plt.title('Cumulative Sales Comparison: Tata vs Mahindra')
plt.xticks(rotation=45)
plt.show()
```

```python
segment_df = df[df['Segment'].str.lower() == 'compact suv']
brand_sales = segment_df.groupby('Brand')['Sales Volume'].sum()

plt.figure(figsize=(6, 6))
plt.pie(brand_sales, labels=brand_sales.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Compact SUV Market Share: Tata vs. Mahindra')
plt.axis('equal')
plt.show()
```

```python
segment_df = df[df['Segment'].str.lower() == 'mid size suv']
brand_sales = segment_df.groupby('Brand')['Sales Volume'].sum()

plt.figure(figsize=(6, 6))
plt.pie(brand_sales, labels=brand_sales.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Mid Size SUV Market Share: Tata vs. Mahindra')
plt.axis('equal')
plt.show()
```
